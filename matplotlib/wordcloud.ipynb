```python
# 라이브러리 로드

import matplotlib.pyplot as plt
import koreanize_matplotlib
import pandas as pd
import numpy as np
from wordcloud import WordCloud
```

```python
text = 'square'

wc = WordCloud(background_color="white", repeat=True)
wc.generate(text)

plt.figure(figsize=(10,6))
plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```

```python
with open("./data/alice.txt") as f:
    text = f.read()

wc = WordCloud(background_color="white", repeat=True)
wc.generate(text)

plt.figure(figsize=(10,6))
plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```

```python
from PIL import Image

with open("./data/alice.txt") as f:
    text = f.read()

# 마스크 이미지 가져오기
# array에 담는 이유 : 숫자코드형식으로 이미지를 담아오기 때문
alist_mask = np.array(Image.open("./data/alice_mask.png"))

wc = WordCloud(background_color="white", repeat=True, mask=alist_mask)
wc.generate(text)

plt.figure(figsize=(10,6))
plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```
