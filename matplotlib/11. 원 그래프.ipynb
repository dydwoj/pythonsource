```python
# 라이브러리 로드

import matplotlib.pyplot as plt
import koreanize_matplotlib
import pandas as pd
import numpy as np
```

#### [실습1]

```python
values = [30,25,20,13,10,2]
labels = ['Python', 'Java', "JavaScript", "C#", "C/C++", "ETC"]
```

```python
g = plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False)
```

- autopct : 원 안에 표시할 % 포맷
- startangle : 안 주는 경우 오른쪽 15도 정도에서 시작해서 시계 반대 방향으로 원이 돌아가게 됨, 각도를 주게 되면 그 각도를 기준으로 시계 반대 방향으로 원이 그려짐
- counterclock=False : 시계 방향으로 원이 작성되도록 함

#### [실습2] 원 조각 내기

```python
explode = [0.2, 0.1, 0, 0, 0, 0]
g = plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False, explode=explode)
```

```python
explode = [0.05] * 6
g = plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False, explode=explode)
```

#### [실습3] 범례

```python
explode = [0.05] * 6
plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False, explode=explode)
g = plt.legend()
```

```python
explode = [0.05] * 6
plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False, explode=explode)
g = plt.legend(loc=(1.1, 0.3), title="언어별 선호도")
```

## 심화

### 1) 색상 변경

```python
explode = [0.05] * 6
colors = ["#ffadad", "#ffd6a5", "#fdffb6", "#caffbf", "#9bf6ff", "#a0c4ff"]
plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False, explode=explode, colors=colors)
g = plt.legend(loc=(1.1, 0.3), title="언어별 선호도")
```

### 2) wedgeprops() : 도넛 모양 형태로 구성

```python
explode = [0.05] * 6
colors = ["#ffadad", "#ffd6a5", "#fdffb6", "#caffbf", "#9bf6ff", "#a0c4ff"]
plt.pie(
    values,
    labels=labels,
    autopct="%.1f%%",
    startangle=90,
    counterclock=False,
    explode=explode,
    colors=colors,
    wedgeprops={"width":0.8},
)
g = plt.legend(loc=(1.1, 0.3), title="언어별 선호도")
```

#### explode 대신에 wedgeprops 이용

```python
colors = ["#ffadad", "#ffd6a5", "#fdffb6", "#caffbf", "#9bf6ff", "#a0c4ff"]
plt.pie(
    values,
    labels=labels,
    autopct="%.1f%%",
    startangle=90,
    counterclock=False,
    colors=colors,
    wedgeprops={"width":0.8, "edgecolor":"lightgray", "linewidth":3},
)
g = plt.legend(loc=(1.1, 0.3), title="언어별 선호도")
```

### 3) autopct 커스텀

```python
def custom_autopct(pct):
    return ("%.1f%%" % pct) if pct >= 10 else ''
```

```python
colors = ["#ffadad", "#ffd6a5", "#fdffb6", "#caffbf", "#9bf6ff", "#a0c4ff"]
plt.pie(
    values,
    labels=labels,
    autopct=custom_autopct,
    startangle=90,
    counterclock=False,
    colors=colors,
    wedgeprops={"width":0.8, "edgecolor":"lightgray", "linewidth":3,},
)
g = plt.legend(loc=(1.1, 0.3), title="언어별 선호도")
```
