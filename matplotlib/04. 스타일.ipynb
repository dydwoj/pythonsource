## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
# 라이브러리 로드

import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

#### 1. 선 굵기 : linewidth

```python
g = plt.plot(x, y, linewidth=8)
```

#### 2. 마커
- https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html


```python
g = plt.plot(x, y, marker='o')
```

```python
# 마커 사이즈 변경

g = plt.plot(x, y, marker='o', markersize=10)
```

```python
g = plt.plot([4,5,6], marker='H')
g = plt.plot([3,4,5], marker='d')
g = plt.plot([2,3,4], marker='x')
g = plt.plot([1,2,3], marker=11)
g = plt.plot([0,1,2], marker='+')
```

```python
# markeredgecolor : 마커 테두리 색상
# markerfacecolor : 마커 색상

g = plt.plot(x,y, marker='o', markersize=10, markeredgecolor="red", markerfacecolor="yellow")
```

```python
g = plt.plot(x,y, markersize=10)
```

#### 3. 선 모양 : linestyle(https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)

```python
g = plt.plot(x,y, linestyle=":", linewidth=5)
```

```python
g = plt.plot(x,y, linestyle="--", linewidth=5)
```

```python
g = plt.plot(x,y, linestyle="-.", linewidth=5)
```

#### 4. 선 색상 : color

- https://matplotlib.org/stable/api/colors_api.html#module-matplotlib.colors

```python
# r : red
# g : green
# b : blue
g = plt.plot(x,y, color="r")
```

```python
g = plt.plot(x,y, color="b")
```

```python
g = plt.plot(x,y, color="g")
```

#### 5. 포맷 - fmt = '[marker][line][color]'

```python
g = plt.plot(x,y, 'o')
```

```python
g = plt.plot(x,y, 'ro')
```

```python
g = plt.plot(x,y, 'ro--')
```

```python
g = plt.plot(x,y, 'bv:')
```

```python
g = plt.plot(x,y, 'go')
```

#### 6. 축약(https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
# mfc : markerfacecolor
# mec : markeredgecolor
# ls : linestyle
# ms : markersize

g = plt.plot(x,y, marker="o", mfc="red", mec="b", ls=":", ms=10)
```

#### 7. 투명도

```python
g = plt.plot(x,y, marker="o", mfc="red", mec="b", ls=":", ms=10, alpha=0.3)
```

#### 8. 그래프 크기

```python
# plt.figure(figsize=())

plt.figure(figsize=(7, 3))
g = plt.plot(x,y, marker="o", mfc="red", mec="b", ls=":", ms=10, alpha=0.3)
```

```python
plt.figure(figsize=(3, 7))
g = plt.plot(x,y, marker="o", mfc="red", mec="b", ls=":", ms=10, alpha=0.3)
```

```python
# dpi : 확대의 의미가 있음
plt.figure(figsize=(10, 5), dpi=200)
g = plt.plot(x,y, marker="o", mfc="red", mec="b", ls=":", ms=10, alpha=0.3)
```

#### 9. 배경색

```python
plt.figure(figsize=(10, 5), facecolor='yellow')
g = plt.plot(x,y)
```
