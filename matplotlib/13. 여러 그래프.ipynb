```python
# 라이브러리 로드

import matplotlib.pyplot as plt
import koreanize_matplotlib
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/student.csv")
df
```

### subplots()

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle("여러개의 그래프")

# 첫번째 (국어점수 : 이름, 점수)
axes[0,0].bar(df['이름'], df["국어"])
axes[0,0].set_title("학생별 국어점수")
axes[0,0].grid(linestyle="--", linewidth=0.5)

# 두번째
axes[0,1].plot(df['이름'], df["영어"])
axes[0,1].plot(df['이름'], df["수학"])
axes[0,1].set_title("학생별 영어/수학 점수")
axes[0,1].legend(["영어", "수학"])

# 세번째
axes[1,0].barh(df['이름'], df["키"])
axes[0,0].set_title("학생별 키")

# 네번째
axes[1,1].plot(df['이름'], df["사회"])
```
