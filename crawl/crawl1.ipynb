### 크롤링 vs 스크랩핑

- 크롤링 : 주기적으로 , 무작위로 데이터 수집하느 것 (데이터의 양을 우선으로 수집)
    - 사이트에서 허용하는 경로만 가능
- 스크랩핑 : 어떤 특정한 카테고리를 선택해서 데이터를 수집 (크롤링보다 작은 개념)

```python
import urllib.request as req

url = "http://google.com"

# urlretrieve
# 요청하는 url 정보를 로컬파일로 저장

file1, header1 = req.urlretrieve(url, "./google.html")
print(header1)
```

```python
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/c/c3/Python-logo-notext.svg/1200px-Python-logo-notext.svg.png"

# urlretrieve
# 요청하는 url 정보를 로컬파일로 저장

file1, header1 = req.urlretrieve(url, "./python.png")
print(header1)
```

```python
import urllib.request as req

url = "http://encar.com"

response = req.urlopen(url)
print(response)

print(f"url {response.geturl()}")
print(f"url {response.status}")
print(f"url {response.getheaders()}") # 튜플 구조로 가져옴

contents = response.read().decode("euc-kr")
print(contents[:4000])
```

### requests
- pip install requests
- urllib.equest 모듈과 같은 역할
- 디코딩도 알아서 해주고, json 처리도 편함

```python
import requests

url = "http://encar.com"

response = requests.get(url)
print(response.status_code)
print(response.headers['content-type'])
print(response.text)
```

```python
# 세션 이용

session = requests.Session()

response = requests.get(url)
print(response.status_code)
print(response.headers['content-type'])
print(response.text)

session.close()
```

```python
# get

url = "https://httpbin.org/get"
requests.get(url)

# get + 파라메터
# ex) ?page=1&size=10 로 만들기 => 딕셔너리 구조로 만들기

params = {"test": "name"}

res = requests.get(url, params=params)
print(res.headers)
print(res.text)
print(res.json())
```

```python
# post

# post 는 대부분 form data 가 같이 따라감 => 딕셔너리 구조로 만들기

data = {"name": "hong"}

res = requests.post("https://httpbin.org/post", data=data)
print(res.text)
```

```python
# 행정안전부 rss
# https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp?ctxCd=1012
# https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp?ctxCd=1013
# https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp?ctxCd=1001

params = []

for code in [1012, 1013, 1001]:
    params.append(dict(ctxCd=code))

print(params)

url = "https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp"

with requests.Session() as s:
    for param in params:
        res = s.get(url, params=param)
        print("-" * 100)
        print(res.text)
```

```python
# 일반 베스트 100
# https://shoppinghow.kakao.com/siso/p/api/bestRank/dispprodbest?vCateId=GMP&durationDays=30&count=100&_=1749693115175

# 컴퓨터 베스트 100
# https://shoppinghow.kakao.com/siso/p/api/bestRank/dispprodbest?vCateId=B02&durationDays=30&count=100&_=1749695851655
# 패션 베스트 100
# https://shoppinghow.kakao.com/siso/p/api/bestRank/dispprodbest?vCateId=A01&durationDays=30&count=100&_=1749695981957

url = "https://shoppinghow.kakao.com/siso/p/api/bestRank/dispprodbest?durationDays=30&count=100"
params = []
cates = ["B02", "A01"]
for cate in cates:
    params.append(dict(vCateId=cate))

ids = [1749695851655, 1749695981957]
for id in ids:
    params.append(dict(_=id))

with requests.Session() as s:
    for param in params:
        res = s.get(url, params=param)
        print("-" * 100)
        print(res.text)
```

```python
# 기사

url = "https://v.daum.net/v/20250613091125816"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

# 기사 제목
name = soup.h3.text
print(f"기사제목 : {name}")

# 기자명
reporter = soup.find("span", class_="txt_info")
print(f"기자명 : {reporter.text}")
# 강사님 기자명
date_time = reporter.find_next_sibling("span")
print(f"강사님 기자명 : {date_time.text}")

# 작성일시
date = soup.find("span", class_="num_date").text
print(f"작성일시 : {date}")

# 첫문단
dialogue = soup.find("div", attrs={"dmcf-ptype":"general"})
d = dialogue.find("div").text.strip()
print(f"첫 문단 : {d}")
```

```python
url = "https://pythonscraping.com/pages/page3.html"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

# Totally Normal Gifts 출력
h1 = soup.h1.text
print(f"제목 : {h1}")

# 상단 본문 내용 출력
content = soup.select_one("#content").text
print(f"상단 본문 내용 : {content}")

# 모든 img 태그 찾기 출력
# src 출력 ['src']
imgs = soup.find_all("img")
for img in imgs:
    print(img)
    print(img['src'])

# 테이블 제목행 출력
gifts = soup.select(".gift")
for t in gifts:
    td = t.select_one("td").text
    print(f"테이블 제목행 : {td.strip()}")

# 선물 가격만 출력
for p in gifts:
    cost = p.select_one("td:nth-child(3)")
    print(f"선물가격 : {cost.text.strip()}")
```
