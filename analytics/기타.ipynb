### Youtube 댓글 크롤링 후 데이터 프레임 생성

````python

# 무한 스크롤
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

from bs4 import BeautifulSoup
````

````python


#콜드플레이 뮤비 아래 달려있는 댓글

#유튜브 링크
url = 'https://www.youtube.com/watch?v=HosW0gulISQ'

#웹드라이버 가져오기
driver = webdriver.Chrome()

driver.get(url)
time.sleep(3)

# 스크롤 이동
# 현재 문서 높이 (스크롤 이동을 위해서)
prev_height = driver.execute_script("return document.documentElement.scrollHeight")
while True:
    # 스크롤 이동
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight)")

    time.sleep(3)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return document.documentElement.scrollHeight")

    if curr_height == prev_height:
        break

    prev_height = curr_height

# 스크롤 이동이 완료된 후 맨 처음으로 스크롤 이동
driver.execute_script("window.scrollTo(0,0)")

# 페이지 소스 파싱
soup = BeautifulSoup(driver.page_source, 'html.parser')
````

````python
# 가져온 내용 확인하기

# 댓글, 작성자명
author_list = soup.select("#author-text > span")
content_list = soup.select("#content-text > span")

print(len(author_list))
print(len(content_list))
````

````python
#가져온 내용 확인하기
for i in range(len(author_list)):
    print(author_list[i].text.strip(), content_list[i].text.strip())
````

````python
# 데이터 정제
ids = []
comments = []

for i in range(len(author_list)):
    cln_id = author_list[i].text.strip()
    cln_id = cln_id.replace("\n", '')
    cln_id = cln_id.replace("@", '')
    cln_id = cln_id.replace(r"\s", '') # \s -> 공백이나 탭을 의미
    ids.append(cln_id)

    cln_content = content_list[i].text.strip()
    cln_content = cln_content.replace("\n", '')
    cln_content = cln_content.replace("\t", '')
    cln_content = cln_content.replace(r"\s", '')
    comments.append(cln_content)
````

````python
#Dataframe 안에 넣어서 쓸수 있게 만들어주기

df = pd.DataFrame({
    "id":ids,
    "comment":comments
})
````

````python
df.to_csv("./data/콜드플레이.csv", index=False)
````

### 타입 변환 : astype() vs to_numeric()

```
to_numeric(arg, errors='raise', downcast=None)

arg : scalar, list, tuple, 1-d array, or Serie
errors{‘ignore’, ‘raise’, ‘coerce’}, default ‘raise’
- 'raise' : 변환할 수 없는 값이면 예외 발생
- 'coerce' : 변환할 수 없는 값이면 NaN 으로
- 'ignore' : 변환할 수 없는 값이면 입력 값 그대로

제공되는 데이터 형태에 맞춰 적절하게 숫자 타입으로 변환해 줌 (float64, int64)
```

````python
ser = pd.Series(["1.0", "2", -3])
ser
````

````python
pd.to_numeric(ser)
````

````python
ser = pd.Series(["apple", "1.0", "2", -3])
ser
````

````python
pd.to_numeric(ser)
````

````python
pd.to_numeric(ser, errors="ignore")
````

````python
pd.to_numeric(ser, errors="coerce")
````

````python
ser = pd.Series([None, "1.0", "2", -3])
ser
````

````python
pd.to_numeric(ser)
````

````python
# 데이터 값에 None 이 있으면 float64 로 가야지 에러가 안남
ser.astype("float64")
````
