## Ⅱ. 자연어 처리(Natural Language Processing)

- 일상생활에서 사용하는 언어의 의미를 분석하여 컴퓨터가 처리할 수 있도록 하는 과정
- **자연어 처리 용어**
  - 말뭉치(corpus) : 자연어 처리에서 모델을 학습시키기 위한 데이터
  - 토큰(token) : 자연어 처리를 위해서 아주 작게 나눈 단위
  - 토큰화(tokenization) : 텍스트를 문장이나 단어로 분리
  - 불용어(stop words) : 문장 내에서 많이 등장하나 분석과 관계 없는 단어(ex she, the, he..)
  - 어간 추출(stemming) : 단어를 기본형태로 만드는 작업
    - ex) consign, consigned, consigning, consignment ==> consign 통일
  - 품사 태깅(part-of-speech tagging) : 주어진 문장에서 품사를 식별하기 위해 붙여주는 태그

## Ⅵ. 자연어 처리로 할 수 있는 일
- 텍스트 분석 : 텍스트 데이터에서 의미 있는 정보를 추출하고 분석
- 기계 번역 : 한 언어의 텍스트를 다른 언어로 변환
- 음성 인식 : 음성 데이터를 텍스트 형태로 변환
- 텍스트 생성 : 주어진 데이터나 문맥에 기반하여 새로운 텍스트를 생성하는 기술
- 감성 분석 : 텍스트 내의 주관적인 정보, 예를 들어 긍정 또는 부정의 감정 분석
