### [퀴즈5]

#### 1) 04-01-2020.csv 파일 dataframe 생성 (데이터프레임 변수 이름은 covid_df 로  지정)

```python
covid_df = pd.read_csv("./data/04-01-2020.csv")
covid_df.head(3)
```

#### 2) covid_df 에서 중복된 Country_Region 각각에 대해 Confirmed 를 모두 더한 값을 컬럼으로 갖는 데이터프레임을 피봇테이블로 만드세요

```python
pd.pivot_table(covid_df, aggfunc='sum', index='Country_Region', values='Confirmed')
```

#### 3) covid_df 에서 중복된 Country_Region 각각에 대해 Confirmed 를 모두 더한 값을 컬럼으로 갖는 데이터프레임을 피봇테이블로  만들고, 총 값을 'Total' 을 인덱스 값으로 하는 record 로 추가하세요
- 결측치는 0으로 처리

```python
pd.pivot_table(
    covid_df,
    aggfunc="sum",
    index="Country_Region",
    values="Confirmed",
    margins_name="total",
    margins=True,
    fill_value=0,
)
```
