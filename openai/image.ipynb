```python
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
```

```python
client = OpenAI()
Model = "dall-e-3"
```

```python
response = client.images.generate(model=Model, prompt="흰색 샴 고양이", size="1024x1024", quality="standard")
```

```python
response
```

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt


def show_image_from_url(url):
    res = requests.get(url)
    img = Image.open(BytesIO(res.content))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
show_image_from_url(response.data[0].url)
```
